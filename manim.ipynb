```python
from manim import *
import numpy as np
```

```python
%%manim SquareToCircle


radii = [1, 1, 2, 1]
offsets = [2, 3, 7, 10]


class SquareToCircle(Scene):
    def construct(self):
        # Creating shapes
        x_range = 15
        y_range = 10
        axis_scale = 0.7
        ax = Axes(
            x_range=[-x_range, 0],
            y_range=[0, y_range],
            x_length=x_range * axis_scale,
            y_length=y_range * axis_scale,
            axis_config={"color": BLUE},
            tips=False,
        )
        
        h = ValueTracker(6)
        
        arcs = []
        arc_closing_lines = []
        lines = []
        sectors = []
        for radius, offset in zip(radii, offsets):
            arcs.append(Arc(
                radius=radius * axis_scale, 
                arc_center=ax.coords_to_point(-offset,0),
                color=BLUE,
            ))
            arc_closing_lines.append(Line(
                start=ax.coords_to_point(-offset,radius),
                end=ax.coords_to_point(-offset,0),
                color=BLUE,
            ))
            lines.append(Line(
                start=ax.coords_to_point(-offset,0),
                end=ax.coords_to_point(0,h),
                color=GREEN,
            ))
            
            angle = np.arctan(h / offset)
            sector = Sector(
                outer_radius=radius * axis_scale, 
                color=YELLOW_C,
                start_angle=angle,
                angle=-angle,
            )
            center = sector.get_center() / axis_scale
            sector.move_to(ax.coords_to_point(-offset + center[0], 0 + center[1]))
            sectors.append(sector)
        
        square = Square(
            side_length=1,
            fill_color=YELLOW,
            fill_opacity=1,
            stroke_opacity=0,
        )
        

        #Showing shapes
        self.play(Create(ax))
        self.play(*(Create(arc) for arc in arcs), *(Create(arc_cl) for arc_cl in arc_closing_lines))
        self.play(*(Create(line) for line in lines))
        self.play(*(Create(sector) for sector in sectors))
        
        self.play(Create(square))
        self.play(Transform(sectors[0], square))
        # self.play(FadeOut(square))
        
        self.wait(2)
```
